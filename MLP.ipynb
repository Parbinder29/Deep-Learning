```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
x_train
# this states that our datasets is not image but is an array
```

```python
x_train.shape
#shows our x is of 3 by 1 matrix
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
# will show us our first dataset as an image rather than array
```

```python
x_train[0]
```

```python
x_train = x_train/255
x_test = x_test/255
# every value of that array will become from 0 to 1
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28))) #flatten layes has input of shape 28*28
model.add(Dense(128,activation='relu'))  #Dense layer using relu loss function
model.add(Dense(10,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="Adam",metrics=["accuracy"])
# defined our loss function and optimizer model
```

```python
history = model.fit(x_train,y_train,epochs=10,validation_split=0.2)
# training 80% and rest testing
```

```python
model.predict(x_test)
y_prob = model.predict(x_test)
```

```python
y_prob
```

```python
y_pred = y_prob.argmax(axis=-1)
# these stores the predicted values
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
