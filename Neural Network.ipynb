```python
x = df.drop(columns=['Exited'])
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train = x_train.drop(columns=['Surname'], errors='ignore')
x_test = x_test.drop(columns=['Surname'], errors='ignore')
#rop(columns=['Surname']) coz it was text data
```

```python
import pandas as pd

x_train = pd.get_dummies(x_train, drop_first=True)
x_test = pd.get_dummies(x_test, drop_first=True)
#This function is used for encoding categorical (text) data into numbers
```

```python
x_train, x_test = x_train.align(x_test, join='left', axis=1, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
#scaling of x is done 
```

```python
import tensorflow
from tensorflow import keras 
from tensorflow.keras import Sequential #Imports the Sequential model
from tensorflow.keras.layers import Dense #Imports Dense layer means: Every neuron is connected to every neuron in the next layer (fully connected layer)
```

```python
model = Sequential() 
model.add(Dense(1,activation='relu', input_dim = 13))
#sequential model is used for training
#first used sigmoid then relu for incresing accuracy
```

```python
model.summary()
#summary of the parameters
```

```python
model.compile(loss="binary_crossentropy",optimizer="Adam")
#binary loss function is used and optimizer Adam is used as it is most efficient
```

```python
model.fit(x_train_scaled, y_train, epochs=15)
#model is trained now
```

```python
model.layers[0].get_weights()
#Will tell us the weights
```

```python
y_log = model.predict(x_test_scaled)
#everything stored in a single variable
```

```python
y_pred = np.where(y_log>0.5,1,0)
#if y_log id greater than 0.5 it willl give 1 elso 0
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
